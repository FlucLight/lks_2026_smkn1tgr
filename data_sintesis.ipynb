```python
import pandas as pd
import numpy as np

file_asli = "umkm_success_prediction_with_success.csv"
try:
    df_asli = pd.read_csv(file_asli)
    print(f" Berhasil memuat data asli: {len(df_asli)} baris.")
except FileNotFoundError:
    print(f" Error: File '{file_asli}' tidak ditemukan! Pastikan script ini ditaruh di folder yang sama dengan datamu, bre.")
    exit()

n_sintetis = 5000
half_n = n_sintetis // 2

np.random.seed(42)

success_sintetis = np.array([1] * half_n + [0] * half_n)

marketing_sintetis = np.zeros(n_sintetis, dtype=int)
advice_sintetis = np.zeros(n_sintetis, dtype=int)
internet_sintetis = np.zeros(n_sintetis, dtype=int)
financial_sintetis = np.zeros(n_sintetis, dtype=int)

for i in range(n_sintetis):
    is_success = success_sintetis[i]
    
    if is_success == 1:
        marketing_sintetis[i] = int(np.clip(np.random.normal(5.8, 0.7), 1, 7))
        advice_sintetis[i] = int(np.clip(np.random.normal(5.5, 0.8), 1, 7))

        internet_sintetis[i] = np.random.choice([1, 0], p=[0.85, 0.15])
        financial_sintetis[i] = np.random.choice([1, 0], p=[0.80, 0.20])
    else:
        marketing_sintetis[i] = int(np.clip(np.random.normal(2.5, 0.9), 1, 7))
        advice_sintetis[i] = int(np.clip(np.random.normal(2.2, 0.8), 1, 7))
        
        internet_sintetis[i] = np.random.choice([1, 0], p=[0.30, 0.70])
        financial_sintetis[i] = np.random.choice([1, 0], p=[0.20, 0.80])

print("Sedang memproses dan menyelaraskan struktur data...")
df_sintetis = pd.DataFrame({
    'Age': np.random.normal(38, 10, n_sintetis).astype(int).clip(18, 65),
    'Education': np.random.choice([1, 2, 3, 4, 5], n_sintetis, p=[0.15, 0.35, 0.30, 0.15, 0.05]),
    'Initial_Capital': np.random.choice([0, 1], n_sintetis, p=[0.60, 0.40]),
    'Financial_Record_Keeping': financial_sintetis,
    'Internet_Usage': internet_sintetis,
    'Business_Plan': np.random.choice([0, 1], n_sintetis, p=[0.40, 0.60]),
    'Marketing_Effort': marketing_sintetis,
    'Partnership': np.random.choice([0, 1], n_sintetis, p=[0.50, 0.50]),
    'Parent_Business_Experience': np.random.choice([0, 1], n_sintetis, p=[0.40, 0.60]),
    'Industry_Experience': np.random.normal(8, 5, n_sintetis).astype(int).clip(0, 25),
    'Owner_Gender': np.random.choice([0, 1], n_sintetis, p=[0.50, 0.50]),
    'Professional_Advice': advice_sintetis,
    'Success': success_sintetis
})

df_final = pd.concat([df_asli, df_sintetis], ignore_index=True)

df_final = df_final.sample(frac=1, random_state=42).reset_index(drop=True)

output_file = "umkm_prediction_succes_final.csv"
df_final.to_csv(output_file, index=False)

print("\n BERHASIL Selesai, Bre!")
print(f"Total dataset sekarang bertambah menjadi: {len(df_final)} baris.")
print(f"File tersimpan lokal dengan nama: '{output_file}'")
```
